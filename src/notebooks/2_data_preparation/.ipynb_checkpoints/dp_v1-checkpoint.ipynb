# 2.0 Data preparation

Drop colums with 1 cardinality

```python
for i in df:
    if df[i].nunique() == 1:
        df = df.drop(columns=i)
```

```python
df.sample(1)
```

Define object type

```python
df['Tendency'] = df['Tendency'].astype(str)
# df['CLASS'] = df['CLASS'].astype(str)
df['NSP'] = df['NSP'].astype(str)
```

Check data distribution

```python
for i in df:
    if df[i].dtype != 'object':
        plt.figure(i)
        plt.title(i)
        df[i].hist()
```
